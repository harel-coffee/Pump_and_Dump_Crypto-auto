....Predict on __training__ set..

```python
print(tn,fp,fn,tp)
```

## HOURLY MODEL

```python
h = pd.read_pickle("OHLCV_data_slide_scaled_h.pkl") 
ohlcv_feat_h = list(pd.read_pickle("ohlcv_features_h.pkl"))
print( "Hourly data",h.shape)
```

```python
# ohlcv_feat_h = pd.read_pickle("ohlcv_dict_h.pkl")
# ohlcv_feat_h
```

```python
rf = RF()
x_train, x_test, y_train, y_test = subsample(h[ohlcv_feat_h], h['pumped_yn'], num=10000)
```

```python
x_train.head()
```

```python
rf = RF()
params_rf = {'n_estimators':[50,100,200]
            ,'max_features': ['auto', 'sqrt', 'log2']
            ,'max_depth' : [10,15,25,50,75]
            ,'criterion' :['gini', 'entropy']}
```

```python
CV_rfc = GridSearchCV(estimator=rf, param_grid=params_rf, cv= 5, scoring="roc_auc")
CV_rfc.fit(x_train, y_train)
```

```python
CV_rfc.best_params_
```

```python
rf_best = RF(criterion = 'entropy', max_depth = 10, max_features='auto', n_estimators=100)
rf_best.fit(x_train,y_train)
```

### Predict on training set...

```python
y_train_prob = rf_best.predict_proba(x_train)[:,1]
y_train_pred = np.zeros_like(y_train)
y_train_pred[y_train_prob > 0.01] = 1
```

```python
tn,fp,fn,tp = confusion_matrix(y_train,y_train_pred).ravel()
print(tn,fp,fn,tp)
```

### Predict on testing set...

```python
y_test_prob = rf_best.predict_proba(x_test)[:,1]
y_test_pred = np.zeros_like(y_test)
y_test_pred[y_test_prob > 0.05] = 1
```

```python
tn,fp,fn,tp = confusion_matrix(y_test,y_test_pred).ravel()
print(tn,fp,fn,tp)
```

```python
del tn,fp,fn,tp
```
